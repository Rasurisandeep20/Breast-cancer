```python
import cv2,os
import numpy as np
from sklearn.model_selection import train_test_split
from keras.utils import np_utils
from keras.models import Sequential
from keras.layers import Dense, Activation, Flatten, Dropout
from keras.layers import Conv2D, MaxPooling2D
from keras.callbacks import ModelCheckpoint
from keras.applications.densenet import DenseNet121
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
data_path = '/content/drive/MyDrive/data_temp/'
categories = os.listdir(data_path)

# Assigning numerical values to categories
label_dict = dict(zip(categories, range(len(categories))))
print(label_dict)
print(categories)
```

```python
img_size = 256
data = []
label = []

# Loop through each category and resize the images
for category in categories:
    folder_path = os.path.join(data_path, category)
    img_names = os.listdir(folder_path)
        
    for img_name in img_names:
        img_path = os.path.join(folder_path, img_name)
        img = cv2.imread(img_path)
        try:
            gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)  
            resized = cv2.resize(gray, (img_size, img_size))
            # Resizing the image into 256 x 256, since we need a fixed common size for all the images in the dataset
            data.append(resized)
            label.append(label_dict[category])
            # Appending the image and the label(categorized) into the list (dataset)
        except Exception as e:
            print('Exception:', e)
            # If any exception raised, the exception will be printed here. And pass to the next image
```

```python
data = np.array(data)/255.0
data = np.reshape(data, (data.shape[0], img_size, img_size, 1))
label = np.array(label)
new_label = np_utils.to_categorical(label)

# Splitting the dataset into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(data, new_label, test_size=0.1)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,10))
for i in range(20):
    plt.subplot(5,5,i+1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(np.squeeze(x_test[i]))
    plt.xlabel(categories[np.argmax(y_test[i])])
plt.show()
```

```python
# Defining the CNN model
cnn_model = Sequential()
cnn_model.add(Conv2D(32, (3, 3), input_shape=data.shape[1:]))
cnn_model.add(Activation('relu'))
cnn_model.add(MaxPooling2D(pool_size=(2, 2)))

cnn_model.add(Conv2D(32, (3, 3)))
cnn_model.add(Activation('relu'))
cnn_model.add(MaxPooling2D(pool_size=(2, 2)))

cnn_model.add(Conv2D(32, (3, 3)))
cnn_model.add(Activation('relu'))
cnn_model.add(MaxPooling2D(pool_size=(2, 2)))

cnn_model.add(Flatten())
cnn_model.add(Dropout(0.2))

cnn_model.add(Dense(128, activation='relu'))
cnn_model.add(Dropout(0.1))
cnn_model.add(Dense(64, activation='relu'))
cnn_model.add(Dense(2, activation='sigmoid'))

cnn_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Defining the DenseNet model
densenet_model = DenseNet121(input_shape=data.shape[1:], include_top=True, weights=None, classes=2)
densenet_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Training the models
cnn_history = cnn_model.fit(x_train, y_train, epochs=10, validation_split=0.2)
```

```python
densenet_history = densenet_model.fit(x_train, y_train, epochs=10, validation_split=0.2)
```

```python
# Plotting the training loss and accuracy for both models
import matplotlib.pyplot as plt
N = 10 #number of epochs
plt.style.use("ggplot")
plt.figure()
plt.plot(np.arange(0, N), cnn_history.history["loss"], label="CNN train_loss")
plt.plot(np.arange(0, N), cnn_history.history["val_loss"], label="CNN val_loss")
plt.plot(np.arange(0, N), cnn_history.history["accuracy"], label="CNN train_acc")
plt.plot(np.arange(0, N), cnn_history.history["val_accuracy"], label="CNN val_acc")
plt.plot(np.arange(0, N), densenet_history.history["loss"], label="DenseNet train_loss")
plt.plot(np.arange(0, N), densenet_history.history["val_loss"], label="DenseNet val_loss")
plt.plot(np.arange(0, N), densenet_history.history["accuracy"], label="DenseNet train_acc")
plt.plot(np.arange(0, N), densenet_history.history["val_accuracy"], label="DenseNet val_acc")
plt.title("Training Loss and Accuracy")
plt.xlabel("Epoch #")
plt.ylabel("Loss/Accuracy")
plt.legend(loc="center right")
plt.savefig("CNN_DenseNet_Model")
```
